```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 假設已經有一個 DataFrame，命名為 df
df = pd.read_feather('Step2.feather')

plt.figure(figsize=(8, 6))
sns.countplot(x='飆股', data=df)
plt.title('Target Variable Distribution')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns[:5]
df[numeric_columns].hist(figsize=(12, 10), bins=20)
plt.suptitle('Histogram of Numeric Features')
plt.show()
```

```python
#Step 5: 散佈圖矩陣（查看數值變數間的關聯）
sns.pairplot(df[numeric_columns].sample(1000))
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()
```

```python
# Step 6: 熱力圖（查看數值變數間的相關性）
corr = df[numeric_columns].sample(1000).corr()  # 計算相關性
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
