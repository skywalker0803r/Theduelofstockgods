```python
import pandas as pd
import numpy as np

def sample_df_like_test(df, test_df, column, n_samples=None, bins=100):
    # Step 1: 建立 test_df 的分布（直方圖）
    counts, bin_edges = np.histogram(test_df[column], bins=bins, density=True)

    # Step 2: 為每筆 df 的資料找對應的 bin
    bin_indices = np.digitize(df[column], bin_edges[:-1], right=True)

    # Step 3: 對每個 bin 計算機率（總和 = 1）
    probs = counts / counts.sum()

    # Step 4: 根據機率加權抽樣
    sample_weights = np.array([probs[i-1] if 0 < i <= len(probs) else 0 for i in bin_indices])

    # 去除權重為 0 的（避免掉不在範圍內的）
    valid_idx = sample_weights > 0
    df_valid = df[valid_idx]
    weights = sample_weights[valid_idx]

    # Step 5: 抽樣
    n_samples = n_samples or len(test_df)
    sampled_df = df_valid.sample(n=n_samples, weights=weights, replace=True, random_state=42)

    return sampled_df
```

```python
# 看看分布像不像：
import seaborn as sns
import matplotlib.pyplot as plt
sampled_dfs = []
for col in x_col:
    sampled_df = sample_df_like_test(df, test_df, column=col, bins=30)
    sns.kdeplot(df[col], label='original df')
    sns.kdeplot(test_df[col], label='test_df')
    sns.kdeplot(sampled_df[col], label='sampled_df')
    sampled_dfs.append(sampled_df)
    plt.legend()
    plt.title("Distribution Matching")
    plt.show()

df_combined = pd.concat(sampled_dfs,axis=0,ignore_index=True)
df_combined
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in x_col:
    plt.figure(figsize=(8, 5))  # 開一張新圖

    sns.kdeplot(data=df_combined, x=i, label='train', fill=True, common_norm=False)
    sns.kdeplot(data=test_df, x=i, label='test', fill=True, common_norm=False)

    plt.title(f'Distribution of {i}')
    plt.show()
```
