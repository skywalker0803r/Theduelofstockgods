```python
import pandas as pd

df = pd.read_feather('Step1_30000.feather')
df
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = df.drop(columns=['ID','飆股'])  # 所有特
X = X.fillna(X.mean())
X = X.dropna(axis=1)
y = df['飆股']

selector = SelectKBest(score_func=f_classif, k=50)  # 選前50個最重要的
X_new = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
```

```python
selected_features
```

```python
df[['ID']+list(selected_features)+['飆股']].to_feather('train_data_v1.feather')
```
